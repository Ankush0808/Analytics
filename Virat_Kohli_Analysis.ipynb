### Perform a Anova test to understand is there a relation between dismissal type and runs scored by Virat Kohli

```python
from scipy.stats import f_oneway
```

```python
Group_caught=filtered_data[filtered_data['Dismissal']=='caught']['Runs_scored']
Group_notout=filtered_data[filtered_data['Dismissal']=='not out']['Runs_scored']
Group_bowled=filtered_data[filtered_data['Dismissal']=='bowled']['Runs_scored']
Group_lbw=filtered_data[filtered_data['Dismissal']=='lbw']['Runs_scored']
Group_runout=filtered_data[filtered_data['Dismissal']=='run out']['Runs_scored']
Group_stumped=filtered_data[filtered_data['Dismissal']=='stumped']['Runs_scored']
Group_hitwicket=filtered_data[filtered_data['Dismissal']=='hit wicket']['Runs_scored']
```

```python
f_oneway(Group_caught,Group_runout,Group_lbw,Group_bowled,Group_hitwicket,Group_stumped,Group_notout)
```
